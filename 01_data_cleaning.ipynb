```python
# 1. Create a "Flag" column (1 if missing, 0 if present)
# This explicitly captures the fraud signal mentioned in the brief
df['is_amount_missing'] = df['transaction_amount'].isnull().astype(int)

# 2. Impute (fill) the missing values so the model doesn't crash
# Filling with 0 is standard practice here since a missing amount functionally
# acts like a zero-dollar ping to the system.
df['transaction_amount'] = df['transaction_amount'].fillna(0)

# Check our work!
print("Remaining nulls in transaction_amount:", df['transaction_amount'].isnull().sum())


def clean_amount_text(val):
    if pd.isna(val):
        return 0.0

    # Convert to string and strip out everything except digits and decimals
    cleaned_str = re.sub(r'[^\d.]', '', str(val))


# Apply the regex cleaner
df['transaction_amount'] = df['transaction_amount'].apply(clean_amount_text)


# Drop the redundant 'amt' shadow column
df = df.drop(columns=['amt'])
```

```python
# Merchant category

df["merchant_category"].value_counts()
```

```python
def clean_merchant_category(val):
    # 1. Handle the 78 missing values we found earlier
    if pd.isna(val):
        return 'Unknown'

    val_str = str(val).strip()

    # 2. Define our clean, valid categories (from your top 10 value counts)
    valid_categories = [
        'Electronics', 'Utilities', 'Travel', 'Clothing', 'Grocery',
        'Fuel', 'Entertainment', 'Healthcare', 'Education', 'Food & Dining'
    ]

    # If it's already a perfect match, just return it
    if val_str in valid_categories:
        return val_str

    # 3. Handle the corrupted tail using substring matching
    val_lower = val_str.lower()

    # We check the first 2 or 3 letters of the corrupted string
    if val_lower.startswith('el'): return 'Electronics'
    if val_lower.startswith('ut'): return 'Utilities'
    if val_lower.startswith('tr') or val_lower.startswith('t#') or val_lower.startswith('t?'): return 'Travel'
    if val_lower.startswith('cl') or val_lower.startswith('c#'): return 'Clothing'
    if val_lower.startswith('gr'): return 'Grocery'
    if val_lower.startswith('fu'): return 'Fuel'
    if val_lower.startswith('en'): return 'Entertainment'
    if val_lower.startswith('he') or val_lower.startswith('h#'): return 'Healthcare'
    if val_lower.startswith('ed'): return 'Education'
    if val_lower.startswith('fo'): return 'Food & Dining'

    # Fallback just in case something completely weird slips through
    return 'Unknown'

# Apply the cleaner to the column
df['merchant_category'] = df['merchant_category'].apply(clean_merchant_category)

# Check your work! It should now be perfectly clean.
print(df['merchant_category'].value_counts())

# Step 2: Apply One-Hot Encoding
# We use drop_first=False to ensure we keep 'merchant_Unknown' as a specific feature
df = pd.get_dummies(df, columns=['merchant_category'], prefix='merchant', drop_first=False, dtype=int)

# Check the new columns to ensure it worked!
print([col for col in df.columns if 'merchant_' in col])
```

```python
# 1. Flag the missing balances (The Fraud Signal)
df['is_balance_missing'] = df['account_balance'].isnull().astype(int)

# 2. Force the column to be numeric (just in case there are hidden text strings)
# We fill missing values with -1 so the model knows it's fundamentally different from a true 0.00 balance
df['account_balance'] = pd.to_numeric(df['account_balance'], errors='coerce').fillna(-1.0)

# 3. Engineer the "Balance Utilization" Feature
# We use np.where to safely divide, avoiding "Division by Zero" errors when the balance is 0.00 or -1.0
df['balance_utilization'] = np.where(
    df['account_balance'] > 0,
    df['transaction_amount'] / df['account_balance'],
    -1.0  # Assign -1 if the balance is 0 or missing
)

# Check the new engineered feature!
print(df[['transaction_amount', 'account_balance', 'balance_utilization']].head())
```

```python
# Count how many unique users are associated with each device
users_per_device = df.groupby('device_id')['user_id'].nunique().to_dict()

# Map this count back to our main dataframe
df['unique_users_on_device'] = df['device_id'].map(users_per_device)

# If a device is missing (NaN), we just assume 1 user to be safe
df['unique_users_on_device'] = df['unique_users_on_device'].fillna(1)
```

```python
import ipaddress
import numpy as np

def check_ip_suspicious(ip):
    # If it is completely missing (NaN), it is suspicious
    if pd.isna(ip) or ip in ["", "NA", "$N/A$", "nan", "NULL"]:
        return 1 # Flag as suspicious (1)

    ip_str = str(ip).strip()

    # Ask Python's official networking library if the IP is mathematically valid
    try:
        ipaddress.ip_address(ip_str)
        return 0 # It is valid, not suspicious (0)
    except ValueError:
        return 1 # It failed validation! Flag as suspicious (1)

# 1. Engineer the "Suspicious IP" Fraud Feature
df['is_ip_suspicious'] = df['ip_address'].apply(check_ip_suspicious)

# 2. Check how many bad/missing IPs we caught!
print("Total suspicious IPs found:", df['is_ip_suspicious'].sum())

# 3. Drop the raw text column (Machine Learning models can't read raw IP strings anyway)
df = df.drop(columns=['ip_address'])
```
